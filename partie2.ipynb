# dispo_alim

```python
Bilan_alim=pd.read_csv("data/ba.csv")
Bilan_alim=Bilan_alim.pivot_table(values="Valeur", index=["Pays","Produit","Année","Code Pays","Code Produit"],columns="Élément",aggfunc='sum')
Bilan_alim.columns
```

```python
#On renomme nos colonnes correctement 
Bilan_alim.columns = ['Aliments pour animaux', 'Autres Utilisations', 
               'Disponibilité alimentaire (Kcal/personne/jour)', 
               'Disponibilité alimentaire en quantité (kg/personne/an)', 
               'Disponibilité de matière grasse en quantité (g/personne/jour)', 
               'Disponibilité de protéines en quantité (g/personne/jour)', 'Disponibilité intérieure', 
               'Exportations - Quantité', 'Importations - Quantité', 'Nourriture', 'Pertes', 
               'Production', 'Semences', 'Traitement', 'Variation de stock']
Bilan_alim=Bilan_alim.reset_index()
Bilan_alim=Bilan_alim.fillna(0)#On remplace nos NA donc valeurs absentes par des 0
Unite=['Aliments pour animaux','Autres Utilisations', 'Disponibilité intérieure',
'Exportations - Quantité','Importations - Quantité', 'Nourriture', 'Pertes',
'Production','Semences', 'Traitement', 'Variation de stock']
for element in Unite : #Passer la nourriture en million déjà exprimé avant 
    Bilan_alim[element]=Bilan_alim[element]*1000000
Bilan_alim.head()#On affiche le haut du tableau pour vérifier le bon fonctionnement de notre précédente commande
```

```python
dispo_alim=Bilan_alim[['Pays','Code Pays','Produit','Code Produit','Année',
                'Disponibilité alimentaire en quantité (kg/personne/an)',
                'Disponibilité alimentaire (Kcal/personne/jour)',
                'Disponibilité de matière grasse en quantité (g/personne/jour)',
                'Disponibilité de protéines en quantité (g/personne/jour)']].copy()
#ici il manque la nature du produit avec deux valeurs possibles, 
```

```python
dispo_alim=dispo_alim.merge(population, left_on=['Pays','Année'], right_on=['pays','annee'])#fusion des colonnes
dispo_alim['Disponibilité alimentaire en quantité (kg/personne/an)']=dispo_alim['Disponibilité alimentaire en quantité (kg/personne/an)']*dispo_alim['population']/1000
dispo_alim.head(n=4)
```

```python
Vegetal=pd.read_csv("data/BA_veget.csv")
Vegetal=Vegetal['Produit'].unique()#liste des vegetaux uniquement 
dispo_alim['origin'] = np.where(dispo_alim['Produit'].isin(Vegetal), 'vegetal', 'animal')#On fait une sorte de if avec numpy pour aller transformer soit en animal soit en vegetal
```

```python
dispo_alim.head(n=4)
```

```python
dispo_alim.columns
```

```python
dispo_alim.columns=['pays','code_pays','produit','code_produit','année',
                    'dispo_alim_tonnes','dispo_alim_kcal_p_j','dispo_mat_gr','dispo_prot',
                    'retirer','retirer2','retirer3','retirer4','origin']

```

```python
dispo_alim.drop(['retirer', 'retirer2','retirer3','retirer4'], axis = 1, inplace = True)
```

```python
dispo_alim.reindex(columns = ['pays', 'code_pays', 'année','produit','code_produit',
                      'origin','dispo_alim_tonnes','dispo_alim_kcal_p_j','dispo_prot','dispo_mat_gra'])
dispo_alim.head(n=3)#Probleme j'ai changé l'ordre mais le head ne le prend pas en compte 
```

SELECT sous_nutrition.pays,sous_nutrition.nb_personnes/population.population AS ratio
FROM sous_nutrition,population
WHERE  sous_nutrition.pays=population.pays and sous_nutrition.annee=population.annee and sous_nutrition.annee='2016'
